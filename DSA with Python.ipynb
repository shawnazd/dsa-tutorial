## Extraction of Digits

```python
n = 5873
num = n

while num > 0:
    digit = num % 10
    print(digit)
    num //= 10
```

## Armstrong Number

```python
num = 153

n = num
total = 0

td = len(str(n))

while n > 0:
    d = n % 10
    total += d ** td
    n //= 10

print(num == total)
```

## Frequency Map

```python
num = [5, 6, 101, 9, 7, 7, 1, 9, 12, 1, 5, 8, 12]

freq_map = {}

for i in range(0, len(num)):
    if num[i] in freq_map:
        freq_map[num[i]] += 1
    else:
        freq_map[num[i]] = 1
        
print(freq_map)
```

## Hash Map

```python
num = [5, 6, 101, 9, 7, 7, 1, 9, 12, 1, 5, 8, 12]

hash_map = {}

for i in range(0, len(num)):
    hash_map[num[i]] = hash_map.get(num[i], 0) + 1

print(hash_map)
```
