# Time Complexity

Time Complexity measures how long an algorithm takes to run as the size of the input grows. It helps us analyze the efficiency of an algorithm and predict its performance on large datasets.

Time complexity is usually expressed using Big-O notation, such as O(1), O(n), and O(n²).

## O(1) – Constant Time Complexity

An algorithm has O(1) time complexity if the execution time does not change with input size.

Example: Accessing a specific element in an array.

```
arr = [10, 20, 30, 40]
print(arr[0])  # Always takes the same time
```

In this case, the operation takes the same amount of time whether the array contains 4 elements or 1,000 elements.

## O(n) – Linear Time Complexity

An algorithm has O(n) time complexity if the execution time increases proportionally with the input size.

Example: Traversing all elements in an array.


```
arr = [10, 20, 30, 40]
for x in arr:
    print(x)
```
If the array has n elements, the loop will execute n times, so the running time grows proportionally with n.

## O(n²) – Quadratic Time Complexity

An algorithm has O(n²) time complexity if the execution time grows with the square of the input size. This often happens when a loop is inside another loop.
Example:

```
for i in range(n):
    for j in range(n):
        print(i, j)
```

If n = 10, the total number of operations is approximately 10 × 10 = 100.

## Best, Average, and Worst Case

* **Best Case (Ω):** Runs fastest; input is ideal.
* **Average Case (θ):** Runs typical/expected time; realistic scenario.
* **Worst Case (O):** Runs slowest; input is worst-case.

---

Table – Linear Search Example

| Case    | Symbol | Meaning               | Example                  |
| ------- | ------ | --------------------- | ------------------------ |
| Best    | Ω      | Minimum time          | Target is first element  |
| Average | θ      | Typical/expected time | Target is middle element |
| Worst   | O      | Maximum time          | Target not in array      |

---

**Python Example:**

```python
arr = [10, 20, 30, 40]
target = 10

for x in arr:
    if x == target:
        break  # Best: first, Average: middle, Worst: not present
```

# Space Complexity

Space complexity measures **how much memory an algorithm needs** to run as the input size increases.

It has **two main parts**:

1. **Input Space** – Memory used to store the **input itself**.

   * Example: An array of size n → uses **O(n)** memory.

2. **Auxiliary Space** – Extra memory used by the algorithm (not counting input).

   * Example: Variables, temporary arrays, recursion stack.

---

Examples

1. **O(1) – Constant Space**

* Uses fixed memory regardless of input size.

```python
x = 5
y = 10
z = x + y
```

2. **O(n) – Linear Space**

* Memory grows linearly with input size.

```python
arr = [0] * n  # Array of size n
```

3. **O(n²) – Quadratic Space**

* Memory grows with the square of input size.

```python
matrix = [[0]*n for _ in range(n)]  # n x n matrix
```

---

Summary Table

| Space Complexity | Meaning          | Example         |
| ---------------- | ---------------- | --------------- |
| O(1)             | Constant memory  | Few variables   |
| O(n)             | Linear memory    | Array of size n |
| O(n²)            | Quadratic memory | n x n matrix    |

# Time Limit Exceeded (TLE) Error


A **Time Limit Exceeded (TLE) Error** occurs when an algorithm **takes longer to run than the allowed time** for a problem, usually in programming contests or online judges.

**Reason:**

* The algorithm is **too slow** for the given input size.
* Often caused by **inefficient algorithms** (high time complexity) or **nested loops** that take too long.

**Example:**
Suppose a problem allows **1 second** to process input `n ≤ 10^5`.

```python id="tle_example"
n = 100000
for i in range(n):
    for j in range(n):
        print(i, j)  # O(n^2) → 10^10 operations → TLE
```

* Here, `O(n²)` is too slow for `n = 100000` → **TLE Error**

---

How to Avoid TLE

1. **Use efficient algorithms** (e.g., binary search O(log n) instead of linear search O(n))
2. **Avoid unnecessary loops**
3. **Optimize code** (use fast input/output, avoid recursion if possible)
4. **Check time complexity** before running on large inputs

---

**Shortcut:**

* **TLE = Your code is correct, but too slow.**

## Extraction of Digits

````python
n = 5873
num = n

while num > 0:
    digit = num % 10
    print(digit)
    num //= 10
````

## Armstrong Number

````python
num = 153

n = num
total = 0

td = len(str(n))

while n > 0:
    d = n % 10
    total += d ** td
    n //= 10

print(num == total)
````

## Frequency Map

````python
num = [5, 6, 101, 9, 7, 7, 1, 9, 12, 1, 5, 8, 12]

freq_map = {}

for i in range(0, len(num)):
  if num[i] in freq_map:
    freq_map[num[i]] += 1
  else:
    freq_map[num[i]] = 1

print(freq_map)

freq_map = {}
````

## Hash Map

````python
num = [5, 6, 101, 9, 7, 7, 1, 9, 12, 1, 5, 8, 12]

hash_map = {}

for i in range(0, len(num)):
  hash_map[num[i]] = hash_map.get(num[i], 0) + 1

print(hash_map)
````
